```python
# Importamos las librer铆as necesarias:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Cargamos la base de datos:

df = pd.read_csv('Students_Social_Media_Addiction.csv')
```

```python
# Creamos la funci贸n del EDA:

def eda_basico(df):

    print('Ejemplo de datos del DF:')
    display(df.head(3))
    display(df.tail(3))
    display(df.sample(3))
    print('________________________________________________________________________________________________________')

    print('N煤mero de Filas:')
    display(df.shape[0])
    print('________________________________________________________________________________________________________')

    print('N煤mero de Columnas:')
    display(df.shape[1])
    print('________________________________________________________________________________________________________')

    print('Informaci贸n de la tabla:')
    display(df.info())
    print('________________________________________________________________________________________________________')

    print('Nombre de las columnas:')
    display(df.columns)
    print('________________________________________________________________________________________________________')

    print('Descripci贸n de los datos num茅ricos:')
    num_cols = df.select_dtypes(include=['number'])
    if not num_cols.empty:
        display(num_cols.describe().T)
    else:
        print('锔 No hay columnas num茅ricas en el DataFrame.')
    print('________________________________________________________________________________________________________')

    print('Descripci贸n de los datos no-num茅ricos:')
    cat_cols = df.select_dtypes(include='object')
    if not cat_cols.empty:
        display(cat_cols.describe(include='object').T)
    else:
        print('锔 No hay columnas categ贸ricas (tipo object) en el DataFrame.')
    print('________________________________________________________________________________________________________')

    print('Saber si hay datos 煤nicos:')
    display(df.nunique())
    print('________________________________________________________________________________________________________')

    print('Que datos son nulos por columnas:')
    display(df.isnull().sum())
    print('________________________________________________________________________________________________________')

    print('Filas duplicadas:')
    total_duplicados = df.duplicated().sum()
    if total_duplicados > 0:
        print(f'cantidad de duplicados: {total_duplicados}')
        print('Primeros duplicados')
        display(df[df.duplicated()].head(3))
    else:
        print('No hay duplicados')
    print('________________________________________________________________________________________________________')

    print(' Columnas constantes (solo 1 valor 煤nico):')
    constantes = df.columns[df.nunique() <= 1]
    if len(constantes) > 0:
        print(f'{len(constantes)} columnas con 1 valor 煤nico:')
        display(constantes)
    else:
        print('No hay columnas constantes')
    print('________________________________________________________________________________________________________')
    
    print(' Valores 煤nicos en columnas categ贸ricas:')
    for col in df.select_dtypes(include='object'):
        print(f' {col}')
        print('-----------------------------')
        print(df[col].unique())
        print('________________________________________________________________________________________________________')

    print('К Tipos de datos por columna:')
    display(df.dtypes.value_counts())
    print('________________________________________________________________________________________________________')
```

```python
eda_basico(df)
```

```python
# Generamos una matriz de correlaci贸n

correlation_matrix = df.select_dtypes(include='number').corr()
correlation_matrix
```

```python
# Visualizamos la matriz de correlaci贸n:

plt.figure(figsize = (15, 10))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5,vmin=-1, vmax=1, mask = np.triu(np.ones_like(correlation_matrix, dtype=bool)))

plt.title('Matriz de Correlaci贸n entre Variables')

plt.show()
```
